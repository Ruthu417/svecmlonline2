#### Line Equation
- Y  = mX + c
- Y = b0 + b1 * X
    - b0 --> Y-Intercept
    - b1 --> Slope

**Slope Formula**

```python
# y=mX+c:
    # X=(y-c)/m
# (Xi - mean(X)) * (Yi - mean(Y))
# b1/X  =   -----------------------------------
#                  (Xi - mean(X)) ^2
```

```python
df['Head_Size'] # x
```

```python
X1 = df['Head_Size']     # Input or Independent
Y1 = df['Brain_Weight']  # output or dependent
```

```python
X1, np.mean(X1)
```

```python
Y1, np.mean(Y1)
```

```python
len(X1), len(Y1)
```

```python
numerator = 0
denominator = 0

for i in range(len(X1)):
    numerator += ( X1[i] - np.mean(X1) ) * ( Y1[i] - np.mean(Y1) )
    denominator += ( X1[i] - np.mean(X1) )**2
    
b1 = numerator / denominator
print(b1)# Slope
```

**Intercept Formula**

```python
# b0 = mean(Y1) - b1 * mean(X1)
```

```python
b0 = np.mean(Y1) - (b1 * np.mean(X1))
b0 # intercept
```

**Line Equation Formula**

```python
# Y = b0 + b1*X
Y_predict = b0 + (b1 * X1) # mX+c
Y_predict
```
